```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the dataset
dataset = pd.read_csv("blast_dataset1.csv")

# Extract parameter names and their corresponding titles
parameters = ['B', 'S', 'H', 'D', 'TI', 'TF', 'L', 'C', 'PF', 'RF', 'X']
titles = ['Burden', 'Spacing', 'Bench Height', 'Hole Diameter', 'Initial Stemming Height',
          'Final Stemming Height', 'Charge Length', 'Charge per Hole', 'Powder Factor',
          'Rock Factor', 'Mean Blast Fragment Size']

# Define a custom color palette
custom_palette = sns.color_palette("muted", n_colors=len(parameters))

# Increase the font size and set the font to Times New Roman
plt.rcParams['font.size'] = 20
plt.rcParams['font.family'] = 'Times New Roman'

# Create violin plots for each parameter
plt.figure(figsize=(14, 10))
for i, param in enumerate(parameters):
    plt.subplot(3, 4, i+1)  # Create a grid of 3x4 plots
    sns.violinplot(y=dataset[param], palette=[custom_palette[i]], inner="box", scale="count")  # Show only box plots
    plt.title(titles[i])
    plt.ylabel('')  # To remove the y-axis label for better visualization
    plt.xticks(rotation=45, ha='right')  # Rotate and align number labels
    plt.tight_layout()

plt.show()
```
